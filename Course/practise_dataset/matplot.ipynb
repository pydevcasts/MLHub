```python
x = [0,1,2,3]
y = [3,8,1,10]
fig, ax = plt.subplots(1,2)
ax[0].plot(x, y)
ax[1].bar(x, y)
fig.suptitle("Figure of 2 plot")
plt.show()
```

```python
x = [0,1,2,3]
y = [3,8,1,10]
fig, (ax1, ax2) = plt.subplots(1,2)
fig.suptitle("fig, ax1 ax2")
ax1.grid(axis="y", color="green", linestyle="--", linewidth=1.5)
ax2.grid(color="green", linestyle="--", linewidth=1.5)
# ax1.plot(x, y, marker="^", mec="r", mfc="g")
ax1.plot(x,y,"o:r")
ax2.bar(x, y)
plt.show()
```

```python
x = [0,1,2,3]
y = [3,8,1,10]
fig,ax = plt.subplots(2)
ax[0].plot(x, y, "r")
ax[1].bar(x, y)
plt.show()
```

```python
x = [0,1,2,3]
y = [3,8,1,10]
fig,ax = plt.subplots(2, 2)
ax[0, 0].plot(x, y)
ax[1, 0].bar(x, y)
ax[0,1].plot(x,y, 'tab:green')
ax[1,1].plot(y,x, "tab:red")
plt.show()
```

```python
c=["r","g", "b", "y"]
plt.bar(x,y,width=0.2, color=c)
plt.barh(x,y, height=0.2)
plt.show()
```
